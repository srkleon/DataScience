#### Categorizing States Based on Per Capita GDP

```python
cols = [col for col in df_states.columns if col[:7] == '2014-15' or col == 'Item_AndhraPradesh' or col == 'S.No._AndhraPradesh' ]
df_states_cat = df_states[cols]
df_states_cat1 = df_states_cat[(df_states_cat ['S.No._AndhraPradesh'].str.len() < 3)]
df_states_cat1 = df_states_cat1.drop('S.No._AndhraPradesh', axis = 1)
df_states_cat1 = df_states_cat1.rename({'Item_AndhraPradesh':'Item'}, axis= 1)
df_states_cat1.set_index(['Item'], inplace = True)
df_states_cat2 = df_states_cat1.transpose()
df_states_cat2["Categories"] = pd.qcut(df_states_cat2['Per Capita GSDP (Rs.)'], [0,0.20,0.5, 0.85, 1], ['C4', 'C3', 'C2','C1'])
#df_states_cat2["Categories"]
df_states_cat2.drop(['TOTAL GSVA at basic prices','Taxes on Products','Subsidies on products', 'Per Capita GSDP (Rs.)'],axis = 1, inplace = True)
df_states_cat2
```

```python
df_states_C1 = df_states_cat2[df_states_cat2['Categories'] == 'C1']
df_states_C2 = df_states_cat2[df_states_cat2['Categories'] == 'C2']
df_states_C3 = df_states_cat2[df_states_cat2['Categories'] == 'C3']
df_states_C4 = df_states_cat2[df_states_cat2['Categories'] == 'C4']
df_states_C1_sum = df_states_C1.iloc[:, :12].sum(axis=0)
df_states_C2_sum = df_states_C2.iloc[:, :12].sum(axis=0)
df_states_C3_sum = df_states_C3.iloc[:, :12].sum(axis=0)
df_states_C4_sum = df_states_C4.iloc[:, :12].sum(axis=0)
statelist_C1 = [state[8:] for state in list(df_states_C1.index)]
statelist_C2 = [state[8:] for state in list(df_states_C2.index)]
statelist_C3 = [state[8:] for state in list(df_states_C3.index)]
statelist_C4 = [state[8:] for state in list(df_states_C4.index)]
print('C1 States are : '+ str(statelist_C1))
print('C2 States are : '+ str(statelist_C2))
print('C3 States are : '+ str(statelist_C3))
print('C4 States are : '+ str(statelist_C4))
 #df_states_C1 #.sum(axis = 0) #df_states_C1.sum(axis = 0) # / df_states_C1['Gross State Domestic Product']
```

```python
df_states_C1_perc = df_states_C1_sum.div(df_states_C1_sum['Gross State Domestic Product'],axis = 'index') * 100
df_states_C1_perc.drop('Gross State Domestic Product', inplace = True)
df_states_C1_perc.sort_values(ascending = False, inplace = True)
#df_states_C1_perc
plt.style.use('ggplot')
x_offset = -0.03
y_offset = 0.02

plt.figure(figsize=(20,10))
ax = df_states_C1_perc.plot.bar( )
plt.title("Percentage contribution of different Sub-Sectors to GDP for C1 States")
ax.set_xlabel('Sub-Sectors')
for p in ax.patches:
    b = p.get_bbox()
    val = "{:.2f}".format(b.y1 + b.y0)        
    ax.annotate(val, ((b.x0 + b.x1)/2 + x_offset, b.y1 + y_offset))
plt.savefig('assignmentfig5.png')
plt.show()

print(df_states_C1_perc)
```

```python
df_states_C2_perc = df_states_C2_sum.div(df_states_C2_sum['Gross State Domestic Product'],axis = 'index') * 100
df_states_C2_perc.drop('Gross State Domestic Product', inplace = True)
df_states_C2_perc.sort_values(ascending = False, inplace = True)

plt.figure(figsize=(20,10))
ax = df_states_C2_perc.plot.bar( )
plt.title("Percentage contribution of different Sub-Sectors to GDP for C2 States")
ax.set_xlabel('Sub-Sectors')
for p in ax.patches:
    b = p.get_bbox()
    val = "{:.2f}".format(b.y1 + b.y0)        
    ax.annotate(val, ((b.x0 + b.x1)/2 + x_offset, b.y1 + y_offset))

plt.savefig('assignmentfig6.png')
plt.show()

print(df_states_C2_perc)
```

```python
df_states_C3_perc = df_states_C3_sum.div(df_states_C3_sum['Gross State Domestic Product'],axis = 'index') * 100
df_states_C3_perc.drop('Gross State Domestic Product', inplace = True)
df_states_C3_perc.sort_values(ascending = False, inplace = True)

plt.figure(figsize=(20,10))
ax = df_states_C3_perc.plot.bar( )
plt.title("Percentage contribution of different Sub-Sectors to GDP for C3 States")
ax.set_xlabel('Sub-Sectors')
ax.set_yticks([0, 5, 10, 15, 20])
for p in ax.patches:
    b = p.get_bbox()
    val = "{:.2f}".format(b.y1 + b.y0)        
    ax.annotate(val, ((b.x0 + b.x1)/2 + x_offset, b.y1 + y_offset))

plt.savefig('assignmentfig7.png')
plt.show()

print(df_states_C3_perc)
```

```python
df_states_C4_perc = df_states_C4_sum.div(df_states_C4_sum['Gross State Domestic Product'],axis = 'index') * 100
df_states_C4_perc.drop('Gross State Domestic Product', inplace = True)
df_states_C4_perc.sort_values(ascending = False, inplace = True)

plt.figure(figsize=(20,10))
ax = df_states_C4_perc.plot.bar( )
plt.title("Percentage contribution of different Sub-Sectors to GDP for C4 States")
ax.set_xlabel('Sub-Sectors')
ax.set_yticks([0, 5, 10, 15, 20])
for p in ax.patches:
    b = p.get_bbox()
    val = "{:.2f}".format(b.y1 + b.y0)        
    ax.annotate(val, ((b.x0 + b.x1)/2 + x_offset, b.y1 + y_offset))

plt.savefig('assignmentfig8.png')
plt.show()

print(df_states_C4_perc)
```

```python
drop_rate = drop_rate[[col for col in drop_rate.columns if ('2014-2015' in col ) or (col =='Level of Education - State')]]
drop_rate.drop('Primary - 2014-2015.1', axis = 1,inplace = True)# There were 2 2014-2015 columns for primary in dataset
#print(drop_rate)
drop_rate.rename({'Level of Education - State':'State'}, axis= 1, inplace = True)
drop_rate.set_index('State', inplace = True)
drop_rate.drop(['A & N Islands','Dadra & Nagar Haveli','Daman & Diu','Delhi','Lakshadweep','Puducherry','Chandigarh'], axis = 0,inplace = True)


drop_rate 
```

```python
pd.options.display.float_format = '{:.2f}'.format
#gsdp_14_15 = gsdp1.iloc[3,:]
#gsdp_14_15.rename({'Items  Description':'State'}, inplace = True)
#gsdp_14_15.drop('Duration', axis= 0 , inplace = True)
#gsdp_state = gsdp_14_15.to_frame()
#gsdp_state.rename({'All_India GDP':'All India','West Bengal1':'West Bengal','Jammu & Kashmir':'Jammu and Kashmir','Uttarakhand':'Uttrakhand', 'Andhra Pradesh ':'Andhra Pradesh','Chhattisgarh':'Chhatisgarh'}, axis = 0 , inplace = True)
#gsdp_state.drop('State', axis = 0 , inplace = True)
#gsdp_state.rename({3:'Total GSDP'}, axis = 1 , inplace = True)
#gsdp_state
df_states_per_capita = df_states1 ['Per Capita GSDP (Rs.)'].to_frame()
df_states_per_capita.index = ['Andhra Pradesh','Arunachal Pradesh','Assam','Bihar','Chhatisgarh','Goa','Gujarat','Haryana','Himachal Pradesh','Jammu and Kashmir','Jharkhand','Karnataka','Kerala','Madhya Pradesh','Maharashtra','Manipur','Meghalaya','Mizoram','Nagaland','Odisha','Punjab','Rajasthan','Sikkim','Tamil Nadu','Telangana','Tripura','Uttar Pradesh','Uttrakhand']
df_states_per_capita
```

```python

drop_rate_merge = drop_rate.merge(df_states_per_capita, how = 'outer', left_index = True, right_index = True)
drop_rate_merge
```

```python
# Lets Drop All india and West bengal as appropriate Data is not available for these.
drop_rate_merge.drop(['West Bengal','All India'], axis = 0,inplace = True)
drop_rate_merge.sort_values(by = 'Per Capita GSDP (Rs.)')
```

```python
from matplotlib import cm as cm
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(111)
cmap = cm.get_cmap('BrBG', 30)
cax = ax.imshow(drop_rate_merge.corr(), interpolation="nearest",cmap=cmap)
ax.grid(True)
labels = ['','Primary - 2014-2015','Upper Primary - 2014-2015', 'Secondary - 2014-2015','Senior Secondary - 2014-2015','Per Capita GSDP']
ax.set_xticklabels(labels,fontsize=7)
ax.set_yticklabels(labels,fontsize=12)
fig.colorbar(cax, ticks=[-0.5,-.25,.0,.25, .5,.75, 1])
plt.savefig('assignmentfig9.png')
plt.show()

drop_rate_merge.corr().style.background_gradient(cmap='BrBG')
```

##### From the above Plot we can form following Hypothesis: <br>
The per capita GDP is Negatively related to Primary, Upper Primary , Secondary drop outs. So the less the no of drop outs in these sectors will lead to a better per Capita GDP. So the Drop out rates should be minimised in Primary, Upper Primary , Secondary Education levels. Government should focus on students completing atleast Secondary level of education for better per capita GDP.
